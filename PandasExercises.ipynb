```python
# Wednesday Exercise 1
import pandas as pd
import re

df = pd.read_csv("./imdb_top_1000.csv")
# getting rid of useless columns
df.drop(columns=['Poster_Link','Overview'], inplace=True)
df.loc[df['Released_Year'] == "PG"]
```

```python
# getting oldest movie
oldest_movie_row = df['Released_Year'].idxmin()
oldest_movie_title = df.iloc[oldest_movie_row]["Series_Title"]
```

```python
# getting newest movie
# newest_movie_row = df['Released_Year'].idxmax()
# newest_movie_row = df.iloc[newest_movie_row]
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce').dropna()
most_recent = df.loc[df['Released_Year'] == df['Released_Year'].max()]
most_recent

```

```python
# Top 10 IMDB Rating Movies

df.sort_values('IMDB_Rating', ascending=False).head(10)
```

```python
# Top Movie for Each Genre
genres = df['Genre'].str.split(', ')
df['Genres'] = genres
df.drop(columns=['Genre'])

Top_movies = df.explode('Genres').sort_values("IMDB_Rating")
Top_movies.groupby('Genres').agg({"IMDB_Rating":['max'],"Series_Title":["last"]})
```

```python
# Director with the most movies
df["Director"].value_counts().idxmax()
```

```python
#Star with the most movies

df["Stars"] = df[["Star1", "Star2", "Star3", "Star4"]].values.tolist()
Top_movies_by_Actors = df.explode("Stars")
Top_movies_by_Actors["Stars"].value_counts().idxmax()
```

```python
# Highest Grossing Movie by Genre
Top_movies['Gross'] = Top_movies['Gross'].str.replace(',','')
Top_movies = Top_movies[Top_movies['Gross'].notna()]
Top_movies['Gross'] = pd.to_numeric(Top_movies["Gross"])
Top_movies.sort_values('Gross',inplace=True)
Top_movies.groupby('Genres').agg({"Gross":['max'],"Series_Title":["last"]})


#Top_movies.sort_values('Gross',inplace=True)
#Top_movies[Top_movies['Gross'].notna()]
#Top_movies.groupby('Genres').agg({"Gross":['max'],"Series_Title":["last"]})
```
